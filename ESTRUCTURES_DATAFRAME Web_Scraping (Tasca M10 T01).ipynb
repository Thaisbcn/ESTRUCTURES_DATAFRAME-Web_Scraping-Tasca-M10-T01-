**Diferencias básicas entre los dos métodos de webscraping:** 

***Beautiful Soup:*** 
- Trabaja con documentos HTML y XML descargados previamente. No interactúa directamente con un navegador web.
- No está vinculado a un navegador en particular y no requiere un navegador web para funcionar.
- Es útil para analizar y extraer información de documentos HTML/XML estáticos.

***Selenium:*** 
- Interactúa directamente con un navegador web, permitiendo la automatización de acciones del usuario, como hacer clic en botones, llenar formularios, etc.
- Se utiliza en combinación con un navegador web. Puede automatizar varios navegadores como Chrome, Firefox, Safari, etc.
- Es útil cuando se necesita interactuar con contenido dinámico, realizar pruebas de usuario automatizadas o realizar scraping de páginas que dependen de JavaScript.

Al realizar web scraping, es esencial comprender y tener en cuenta diversos elementos HTML que son comunes en las páginas web. 

    <div> (División): La etiqueta <div> se utiliza para dividir el contenido en bloques o contenedores. Puede contener otros elementos y se usa comúnmente para estructurar el diseño de la página.   

    <a> (Enlace): La etiqueta <a> se utiliza para crear enlaces. Puede contener texto u otros elementos, y tiene un atributo href que indica la URL a la que el enlace apunta.

    <p> (Párrafo): La etiqueta <p> se utiliza para definir párrafos de texto. Es comúnmente utilizada para separar y estructurar el contenido textual.

    <h1>, <h2>, ..., <h6> (Encabezados): Estas etiquetas se utilizan para definir encabezados de diferentes niveles, desde el título principal <h1> hasta subencabezados más pequeños <h2>, <h3>, etc.

    <ul>, <ol>, <li> (Listas): Se utilizan para crear listas no ordenadas (<ul>) u ordenadas (<ol>). Cada elemento de la lista se define con la etiqueta <li>.

    <table>, <tr>, <td>, <th> (Tablas): Se utilizan para representar datos tabulares. <table> define la tabla, <tr> define una fila, <td> define una celda de datos y <th> define una celda de encabezado.

    <img> (Imagen): La etiqueta <img> se utiliza para incrustar imágenes en una página web. Tiene un atributo src que especifica la URL de la imagen.

    <form>, <input>, <button> (Formularios): Estas etiquetas se utilizan para crear formularios en una página web. <form> define el formulario, <input> se utiliza para los campos de entrada, y <button> crea botones.

    <span> (Span): Como mencionaste, la etiqueta <span> se utiliza para aplicar estilos o realizar manipulaciones específicas a una sección de texto.

    <iframe> (Iframe): Se utiliza para incrustar otro documento HTML dentro de un documento principal. Puedes encontrarte con iframes que cargan contenido de otras fuentes.

***1.1A.- WEBSCRAPING CON BeautifulSoup***

```python
def scrape_quotes_selenium(url):
    # Inicializar el navegador Chrome (puedes cambiar a otro navegador si es necesario)
    driver = webdriver.Chrome()

    # Navegar a la URL
    driver.get(url)

    # Encontrar todas las citas en la página utilizando Selenium
    quotes = driver.find_elements(By.CLASS_NAME, "quote")

    # Crear listas para almacenar los datos
    frases = []
    autores = []
    etiquetas_list = []

    # Iterar sobre las citas y extraer la frase, el autor y las etiquetas
    for quote in quotes:
        frase = quote.find_element(By.CLASS_NAME, "text").text
        autor = quote.find_element(By.CLASS_NAME, "author").text
        etiquetas = [tag.text for tag in quote.find_elements(By.CLASS_NAME, "tag")]

        # Agregar datos a las listas
        frases.append(frase)
        autores.append(autor)
        etiquetas_list.append(etiquetas)

    # Cerrar el navegador después de la extracción de datos
    driver.quit()

    # Crear un DataFrame
    df_quotes = pd.DataFrame({
        'Frase': frases,
        'Autor': autores,
        'Etiquetas': etiquetas_list
    })

    # Imprimir el DataFrame
    print(df_quotes)

    # También puedes devolver el DataFrame si es necesario para futuras manipulaciones
    return df_quotes

# URL de la página con citas
url_quotes = "http://quotes.toscrape.com"
scrape_quotes_selenium(url_quotes)
```

```python
df_quotes = pd.DataFrame(columns=['text', 'author', 'tags'])

for i in range(1,11):
    url="https://quotes.toscrape.com/page/" + str(i) + "/"
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    quotes=soup.find_all("div", class_="quote")
    for quote in quotes:
        text=quote.find("span", class_="text").text
        author=quote.find("small", class_="author").text
        tag_html=quote.find_all("a", class_="tag")
        tag_list=[]
        for tag in tag_html:
            tag_list.append(tag.text)
        df_quotes = pd.concat([df_quotes, pd.DataFrame([{'text': text, 'author':author, 'tags': tag_list}])], ignore_index=True)
    
df_quotes.info()
```

**EXERCICI 1.2A.- WEBSCRAPING CON SELENIUM DE LA PÁGINA QUOTES**

```python
from selenium.webdriver import Chrome
from selenium.webdriver.chrome.options import Options
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
```

```python
url_quotes_s= "http://quotes.toscrape.com"
opts = Options()
opts.add_argument('-headless=new')
browser = Chrome(options=opts)
browser.get(url_quotes_s)
```

```python
def scrape_quotes_with_selenium(url):
    # Configurar las opciones de Chrome (opciones headless)
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # Comentar esta línea si deseas ver el navegador en acción

    # Inicializar el navegador Chrome
    driver = webdriver.Chrome(options=chrome_options)

    # Navegar a la URL
    driver.get(url)

    # Encontrar todas las citas en la página utilizando Selenium
    quotes = driver.find_elements(By.CLASS_NAME, "quote")

    # Crear listas para almacenar los datos
    frases = []
    autores = []
    etiquetas_list = []

    # Iterar sobre las citas y extraer la frase, el autor y las etiquetas
    for quote in quotes:
        frase = quote.find_element(By.CLASS_NAME, "text").text
        autor = quote.find_element(By.CLASS_NAME, "author").text
        etiquetas = [tag.text for tag in quote.find_elements(By.CLASS_NAME, "tag")]

        # Agregar datos a las listas
        frases.append(frase)
        autores.append(autor)
        etiquetas_list.append(etiquetas)

    # Cerrar el navegador después de la extracción de datos
    driver.quit()

    # Crear un DataFrame
    df_quotes = pd.DataFrame({
        'Frase': frases,
        'Autor': autores,
        'Etiquetas': etiquetas_list
    })

    # Guardar el DataFrame en un archivo CSV
    df_quotes.to_csv('quotes_data_selenium.csv', index=False)

    # También puedes devolver el DataFrame si es necesario para futuras manipulaciones
    return df_quotes

# URL de la página con citas
url_quotes = "http://quotes.toscrape.com"
scrape_quotes_with_selenium(url_quotes_s)
```

Extrec una taula del link indicat, amb Beautifulsoup

```python
url_smi = "https://datosmacro.expansion.com/mercado-laboral/salario-medio"

try:
    # Realizar una solicitud HTTP para obtener el contenido de la página
    response = requests.get(url_smi)
    response.raise_for_status()  # Lanza una excepción para códigos de respuesta no exitosos

    # Obtener el contenido HTML de la respuesta
    html_content = response.text

    # Crear un objeto BeautifulSoup para analizar el HTML
    soup = BeautifulSoup(html_content, 'html.parser')

    # Buscar todas las tablas en la página
    all_tables = soup.find_all("table")

    # Lista para almacenar los datos extraídos
    data = []

    # Iterar sobre las tablas encontradas
    for table in all_tables:
        # Buscar filas dentro de cada tabla
        rows = table.find_all("tr")

        # Iterar sobre las filas
        for row in rows:
            # Buscar celdas dentro de cada fila
            cells = row.find_all(["th", "td"])  # Cambiado para incluir celdas de encabezado

            # Obtener el contenido de cada celda y agregarlo a la lista de datos
            cell_contents = [cell.text.strip() for cell in cells]
            data.append(cell_contents)

except requests.exceptions.HTTPError as errh:
    print(f"Error HTTP: {errh}")
except requests.exceptions.ConnectionError as errc:
    print(f"Error de conexión: {errc}")
except requests.exceptions.Timeout as errt:
    print(f"Error de tiempo de espera: {errt}")
except requests.exceptions.RequestException as err:
    print(f"Error general de solicitud: {err}")
except Exception as e:
    print(f"Otro error: {e}")

# Convertir la lista de datos en un DataFrame de pandas
df_smi_b = pd.DataFrame(data)
```
